# 模块 A：智能图像处理应用开发

基于 OpenCV 的基础图像处理操作，涵盖图像读取与显示、颜色空间转换、图像滤波、边缘检测、阈值分割等核心功能。

## 第 1 题：导入 OpenCV 库

```python
import cv2
import matplotlib.pyplot as plt
```

## 第 2 题：使用 OpenCV 读取图像 butterfly.jpg，并赋值给变量 image，图像在 data 文件夹中

```python
image = cv2.imread('data/butterfly.jpg')
```

## 第 3 题：将读取的 BGR 格式图像 image 转换为 RGB 格式，并用 matplotlib 展示原始图像

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

## 第 4 题：将彩色图像 image 转换为灰度图像 gray_image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8, 6))
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()
```

## 第 5 题：使用高斯模糊对灰度图像 gray_image 进行平滑处理，生成模糊后的图像 blurred。高斯核大小设置为 5x5，自动计算高斯核在 X 方向的标准差

```python
blurred = cv2.GaussianBlur(gray_image, (5, 5), 0)
plt.figure(figsize=(8, 6))
plt.imshow(blurred, cmap='gray')
plt.axis('off')
plt.title('Blurred Image (Gaussian Blur 5x5)')
plt.show()
```
