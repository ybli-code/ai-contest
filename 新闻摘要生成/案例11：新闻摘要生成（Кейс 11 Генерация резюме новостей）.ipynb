第94题：使用nltk库的函数将长文本full_text按句子边界分割成句子列表（即句子分词），结果存储在变量sentences中。

Задача 94: Используя функции библиотеки NLTK, разделить длинный текст full_text на список предложений (сегментация текста по границам предложений) и сохранить результат в переменной sentences.

```python
full_text = ' '.join(df['article_text'].tolist())
######上传此部分代码(Upload the Code)######
sentences = sent_tokenize(full_text)
######上传此部分代码(Upload the Code)######
print(f"\n总句子数: {len(sentences)}")
print("前3个句子:")
for i, sent in enumerate(sentences[:3]):
    print(f"{i+1}. {sent[:100]}...")
```

```python
def preprocess(sentence):
    sentence = re.sub(r'[^a-zA-Z]', ' ', sentence)
    sentence = re.sub(r'\s+', ' ', sentence)
    words = word_tokenize(sentence.lower())
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)
clean_sentences = [preprocess(sent) for sent in sentences]
```

第95题：使用 CountVectorizer 将清洗后的句子列表 clean_sentences 转换为词袋模型的稀疏矩阵，再计算所有句子两两之间的余弦相似度，生成相似度矩阵 similarity_matrix。   
注意：本题须同时上传代码和运行结果。  
Задача 95: Используйте CountVectorizer для преобразования очищенного списка предложений clean_sentences в разреженную матрицу модели «мешок слов», затем вычислите косинусную сходство между всеми парами предложений, чтобы сгенерировать матрицу сходства similarity_matrix.

```python
from sklearn.feature_extraction.text import CountVectorizer
######上传此部分代码(Upload the Code)######
vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(clean_sentences)
similarity_matrix = cosine_similarity(vectors)
######上传此部分代码(Upload the Code)######
print(f"\n相似度矩阵形状: {similarity_matrix.shape}")
```

```python
nx_graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(nx_graph)
```

第96题：使用generate_summary函数根据句子得分scores。从sentences中选3个句子，生成摘要summary并返回索引selected_indices。  
Question 96: Use the generate_summary function to select 3 sentences from sentences based on their scores (scores), generate a summary (summary), and return the corresponding indices (selected_indices).

```python
def generate_summary(sentences, scores, num_sentences=3):
    ranked_sentences = sorted(((scores[i], i, s) for i, s in enumerate(sentences)), reverse=True)
    selected_indices = sorted([ranked_sentences[j][1] for j in range(num_sentences)])
    summary = ' '.join([sentences[i] for i in selected_indices])
    return summary, selected_indices
######上传此部分代码(Upload the Code)######
summary, selected_indices = generate_summary(sentences, scores, num_sentences=3)
######上传此部分代码(Upload the Code)######

print("\n" + "="*60)
print("生成的摘要:")
print("="*60)
print(summary)
print("="*60)
print(f"使用的句子索引: {selected_indices}")
```

```python
def evaluate_summary(original_text, summary_text):
    original_words = set(preprocess(original_text).split())
    summary_words = set(preprocess(summary_text).split())
    
    coverage = len(summary_words.intersection(original_words)) / len(original_words) if original_words else 0
    compression_ratio = len(summary_text) / len(original_text) if original_text else 0
    
    return {
        'coverage': round(coverage, 3),
        'compression_ratio': round(compression_ratio, 3),
        'summary_length': len(summary_text),
        'original_length': len(original_text)
    }
```
